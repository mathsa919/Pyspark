```python
import os
os.environ["SPARK_HOME"] = "/content/spark-3.1.1-bin-hadoop3.2"
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create DataFrame from CSV File in PySpark 3.1").getOrCreate()
import select
from pyspark.sql.functions import col
```

```python
df1 = "/content/BM/customers.csv"

df_customers = spark.read.csv(path= df1 , sep=",", header=True)

df_customers.show()
```

```python
df_customers.sort("Name","City").show(truncate=False)
```

```python
df_customers.sort(col("City"),col("Address")).show(truncate=False)
```

```python
df_customers.orderBy("Name","Address").show(truncate=False)
```

```python
df_customers.orderBy(col("City"),col("Address")).show(truncate=False)
```

```python
df_customers.sort(df_customers.Name.asc(),df_customers.City.asc()).show(truncate=False)
```

```python
df_customers.sort(col("City").asc(),col("Name").asc()).show(truncate=False)
```

```python
df_customers.orderBy(col("ID").asc(),col("City").asc()).show(truncate=False)
```

```python
df_customers.sort(df_customers.Name.desc(),df_customers.City.desc()).show(truncate=False)
```

```python
df_customers.sort(col("City").desc(),col("Name").desc()).show(truncate=False)
```

```python
df_customers.orderBy(col("ID").desc(),col("City").desc()).show(truncate=False)
```

```python
df_customers.sort(df_customers.Name.asc(),df_customers.City.desc()).show(truncate=False)
```

```python
df_customers.sort(col("Name").asc(),col("City").desc()).show(truncate=False)
```

```python
df_customers.orderBy(col("ID").asc(),col("City").desc()).show(truncate=False)
```
