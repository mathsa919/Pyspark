```python
import os
os.environ["SPARK_HOME"] = "/content/spark-3.1.1-bin-hadoop3.2"
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create DataFrame from CSV File in PySpark 3.1").getOrCreate()
import select
from pyspark.sql.functions import col
from pyspark.sql.functions import lit
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
```

```python
df1 = "/content/BM/products.csv"

df_products = spark.read.csv(path= df1 , sep=",", header=True)

df_products.show()
```

```python
df2 = df_products.dropDuplicates()
print("Distinct count: "+str(df2.count()))
df2.show(truncate=False)
```

```python
dropDisDF = df_products.dropDuplicates(["Price","Rating"])
print("Distinct count of Price & Rating : "+str(dropDisDF.count()))
dropDisDF.show(truncate=False)
```

```python
dropDisDF = df_products.dropDuplicates(["Category","Ean"])
print("Distinct count of Category & Ean : "+str(dropDisDF.count()))
dropDisDF.show(truncate=False)
```

```python
df_products.withColumn("Title",col("Title").cast("String")).show()
```

```python
df_products.withColumn("Rating",col("Rating").cast("Integer")).show()
```

```python
df_products.withColumn("Rating",col("Rating")*100).show()
```

```python
df_products.withColumn("CopiedColumn",col("Rating")* -1).show()
```

```python
df_products.withColumn("CopiedColumn",col("Ean")* -1).show()
```

```python
df_products.withColumn("Category", lit("Widget")).show()
df_products.withColumn("Category", lit("Widget")).withColumn("anotherColumn",lit("anotherValue")).show()
```

```python
df_products.withColumnRenamed("ID","SR.No").show(truncate=False) 
```

```python
df_products.drop("Created At").show() 
```

```python
df_products.drop("Created At","Category","Price").show() 
```

```python
windowSpec  = Window.partitionBy("Price").orderBy("Rating")
df_products.withColumn("row_number",row_number().over(windowSpec)).show(truncate=False)
```

```python
"""rank"""
from pyspark.sql.functions import rank
df_products.withColumn("rank",rank().over(windowSpec)).show()
```

```python
"""dens_rank"""
from pyspark.sql.functions import dense_rank
df_products.withColumn("dense_rank",dense_rank().over(windowSpec)).show()
```

```python
""" percent_rank """
from pyspark.sql.functions import percent_rank
df_products.withColumn("percent_rank",percent_rank().over(windowSpec)).show()
```

```python
"""ntile"""
from pyspark.sql.functions import ntile
df_products.withColumn("ntile",ntile(2).over(windowSpec)).show()
```

```python
""" cume_dist """
from pyspark.sql.functions import cume_dist    
df_products.withColumn("cume_dist",cume_dist().over(windowSpec)).show()
```

```python
"""lag"""
from pyspark.sql.functions import lag    
df_products.withColumn("lag",lag("Rating",2).over(windowSpec)).show()
```

```python
"""lag"""
from pyspark.sql.functions import lag    
df_products.withColumn("lag",lag("Ean",2).over(windowSpec)).show()
```

```python
"""lead"""
from pyspark.sql.functions import lead    
df_products.withColumn("lead",lead("Rating",2).over(windowSpec)).show()
```

```python
windowSpecAgg  = Window.partitionBy("Category")
from pyspark.sql.functions import col,avg,sum,min,max,row_number 
df_products.withColumn("row",row_number().over(windowSpec)) \
  .withColumn("avg", avg(col("Rating")).over(windowSpecAgg)) \
  .withColumn("sum", sum(col("Rating")).over(windowSpecAgg)) \
  .withColumn("min", min(col("Rating")).over(windowSpecAgg)) \
  .withColumn("max", max(col("Rating")).over(windowSpecAgg)) \
  .where(col("row")==1).select("Category","avg","sum","min","max") \
  .show()
```
