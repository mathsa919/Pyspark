```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create DataFrame from CSV File in PySpark 3.1").getOrCreate()
import select
from pyspark.sql.functions import col
from pyspark.sql.functions import lit
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql.types import *
from pyspark.sql.functions import *
from pyspark import SparkContext
import pyspark.sql 
sqlContext = SQLContext
```

```python
dfPath = "/content/BM/orders.csv"
df  = spark.read.csv(path=dfPath  , sep=",", header=True)
df.createOrReplaceTempView("people")
tempview=spark.sql("select Discount,Quantity,Tax from people")
tempview.show()
```

```python
dfPath = "/content/BM/orders.csv"
df  = spark.read.csv(path=dfPath  , sep=",", header=True)
df.createOrReplaceTempView("orders")
tempview=spark.sql("select * from orders").show()
```

```python
dfPath = "/content/HM/products.csv"
df  = spark.read.csv(path=dfPath  , sep=",", header=True)
df.createOrReplaceTempView("products")
tempview=spark.sql("select * from products")
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,Quantity,Tax,Ean,Price,Rating  FROM orders INNER JOIN products ON orders.ID = products.ID");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Rating,Tax,Quantity  FROM orders LEFT JOIN products ON orders.ID = products.ID  ORDER BY products.Rating");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Rating,Tax,Quantity  FROM orders LEFT JOIN products ON orders.ID = products.ID  ORDER BY products.Rating DESC");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Rating,orders.Tax,orders.Quantity  FROM orders RIGHT JOIN products ON orders.ID = products.ID  ORDER BY orders.Quantit");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Rating,orders.Tax,orders.Quantity  FROM orders RIGHT JOIN products ON orders.ID = products.ID  ORDER BY orders.Quantity DESC");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Category,orders.Discount,orders.Quantity  FROM orders FULL OUTER JOIN products ON orders.ID = products.ID  ORDER BY products.Category");
tempview.show()
```

```python
tempview=spark.sql("SELECT orders.ID,products.Category,orders.Discount,orders.Quantity  FROM orders FULL OUTER JOIN products ON orders.ID = products.ID  ORDER BY products.Category DESC");
tempview.show()
```
