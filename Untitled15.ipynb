```python
import os
os.environ["SPARK_HOME"] = "/content/spark-3.1.1-bin-hadoop3.2"
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Create DataFrame from CSV File in PySpark 3.1").getOrCreate()
import select
```

```python
df2 = "/content/HM/products.csv"

df_products = spark.read.csv(path= df2 , sep=",", header=True)

df_products.show()
```

```python
df_products.groupBy("Price").count().show()
```

```python
df_products.groupBy("Category").count().show()
```

```python
df_products.groupBy("Rating").count().show()
```

```python
df3 = "/content/BM/customers.csv"

df_customers = spark.read.csv(path= df3 , sep=",", header=True)

df_customers.show()
```

```python
df_customers.groupBy("Name").count().show()
```

```python
df_customers.groupBy("Zip").count().show()
```

```python
df_customers.groupBy("Source").count().show()
```

```python
df4 = "/content/MM/orders.csv"

df_orders = spark.read.csv(path= df4 , sep=",", header=True)

df_orders.show()
```

```python
df_orders.groupBy("Product ID").count().show()
```

```python
df_orders.groupBy("Quantity").count().show()
```

```python
df_orders.groupBy("Discount").count().show()
```
